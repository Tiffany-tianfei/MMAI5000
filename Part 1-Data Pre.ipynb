```python
#Extract: Questions for all, Facial recognition, questioner demographic Columns
df2 = df.iloc[:, 0:27].join(df.iloc[:, 50:77]).join(df.iloc[:, 102:])
```

```python
# Delete useless column Form Type
# INTERVIEW_START_W99, INTERVIEW_END_W99, FORM_W99 
#df2.drop(columns = 'INTERVIEW_START_W99', inplace = True)
#df2.drop(columns = 'INTERVIEW_END_W99', inplace = True)
df2.drop(columns = 'FORM_W99', inplace = True)
df2.head()
```

```python
#Fill empty values
df2['F_HISP_ORIGIN'].fillna('Not HISP', inplace=True)
df2['F_PARTYLN_FINAL'].fillna('Unknown', inplace=True)
df2['F_BORN'].fillna('Unknown', inplace=True)
df2['F_REG'].fillna('Unknown', inplace=True)
df2['F_INTFREQ'].fillna('Unknown', inplace=True)



missing_values = df2.isnull().sum()

# View the missing value counts for each column
with pd.option_context('display.max_rows', None):
    print("Number of missing values in each column:")
    print(missing_values)
```

```python
# Iterate through each column in the dataframe
for col in df2.columns:
        # Fit and transform the column using the LabelEncoder
        df2[col] = le.fit_transform(df2[col])
        print(list(le.classes_), '\n')

# The dataframe is now label encoded
```

```python
#Check unique value
df_processed = df2.copy()

unique_values_dict = {}
columns_with_single_occurrence_values = []

# Iterate through each column in the DataFrame
for col in df_processed.columns:
    # Get unique values and their counts for the column
    unique_values = df_processed[col].value_counts()
    unique_values_dict[col] = unique_values
    # Check if any value occurs only once in the column
    single_occurrences = unique_values[unique_values == 1]
    if not single_occurrences.empty:
        columns_with_single_occurrence_values.append((col, single_occurrences.index.tolist()))

# Create a DataFrame to display the count of unique values for each column
unique_values_count = {col: len(values) for col, values in unique_values_dict.items()}
unique_values_df = pd.DataFrame(list(unique_values_count.items()), columns=['Column', 'Unique Values Count'])

# Display the DataFrame showing the count of unique values for each column
print("Unique Values Count DataFrame:")
print(unique_values_df)

# Display columns with values that occur only once
if columns_with_single_occurrence_values:
    print("\nColumns with Values Occurring Only Once:")
    for col, values in columns_with_single_occurrence_values:
        print(f"Column: {col}, Values: {values}")
else:
    print("\nNo columns have values that occur only once.")
```
